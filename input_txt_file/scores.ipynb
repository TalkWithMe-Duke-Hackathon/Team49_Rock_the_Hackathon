```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
pd.set_option("mode.copy_on_write", True)

scores = pd.read_csv("/Users/castnut/Desktop/input_txt_file/scores.csv")
```

```python
scores.head(10)
```

```python
# Melt the DataFrame to make it suitable for seaborn lineplot
df_melted = scores.melt(id_vars="Index", var_name="ScoreType", value_name="Score")

# Plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_melted, x="Index", y="Score", hue="ScoreType", marker="o")
plt.title("Trend of Improvement across Different Scores")
plt.xlabel("Reading Times")
plt.ylabel("Score")
plt.legend(title="Score Type")
plt.grid(True)
plt.show()
```

```python
# Create a dictionary to map the actual labels to desired legend labels
label_dict = {
    "AccuracyScore": "Accuracy",
    "FluencyScore": "Fluency",
    "CompletenessScore": "Completeness",
    "PronScore": "Pronunciation",  # This is the label you want for the legend
}

# Rename the 'ScoreType' column values using the label_dict
df_melted["ScoreType"] = df_melted["ScoreType"].map(label_dict)

plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df_melted,
    x="Index",
    y="Score",
    hue="ScoreType",
    style="ScoreType",
    palette="tab10",
    markers=True,
    dashes=False,
)
plt.title("Trend of Improvement across Different Scores", fontsize=14)
plt.xlabel("Reading Times", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# The legend will automatically use the new, more readable names
plt.legend(title="Score Type", fontsize="small", title_fontsize="medium")

plt.grid(True)
plt.show()
```
